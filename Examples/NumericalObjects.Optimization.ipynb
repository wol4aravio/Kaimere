# Modified Hybrid Random Search

### Configuration selection

```python
ars_1_config = {
    'AdaptiveRandomSearch': {
        'init_radius': 1.0,
        'factor_small': 1.1,
        'factor_huge': 1.5,
        'frequency': 5,
        'max_no_change': 5
    }
}
rs_w_sag_1_config = {
    'RandomSearchWithStatisticalAntiGradient': {
        'radius': 1.0,
        'number_of_samples': 5
    }
}
lj_1_config = {
    'LuusJaakolaOptimization': {
        'init_radius': 1.0,
        'number_of_samples': 5,
        'reduction_coefficient': 0.95,
        'recover_coefficient': 0.97,
        'iteration_per_run': 5
    }
}

ars_1 = create_algorithm_from_json(ars_1_config)
rs_w_sag_1 = create_algorithm_from_json(rs_w_sag_1_config)
lj_1 = create_algorithm_from_json(lj_1_config)

mt_1 = MaxTimeTerminator.from_json({
    'MaxTimeTerminator': {
        'max_time': 's:5'
    }
})

config = ModifiedHybridRandomSearch(algorithms=[ars_1, rs_w_sag_1, lj_1], terminators=[mt_1, mt_1, mt_1]).to_json()

print(json.dumps(config, indent=2))
```

### Benchmarking

```python
algorithm = OptimizationTools.create_algorithm_from_json(config)
```

```python
json.dump(config, open('../temp.json', 'w'))

!cd .. && pipenv run benchmark -A temp.json -W s:10 -T Tasks/OptimizationBenchmarks/2D -R 100 -O results --parallel 8

!rm ../temp.json
```

```python
!cat ../results/statistics.md
```

```python
!rm -r ../results/
```
